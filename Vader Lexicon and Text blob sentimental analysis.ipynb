```python
## so applying sent.polarity_scores on full amazon dataset 
```

```python
score_com = []
score_pos = []
score_neg = []


for i in range( 0 , amazon.shape[0] ):
    score = sent.polarity_scores(amazon.iloc[i][1]) ## this [1] is for the second column 
    score1 = score["compound"]
    score_com.append(score1)

    
    ## below 2 lines to extract the positive sentiment 
    
    score2 = score["pos"]
    score_pos.append(score2)
    
    
    ## below 2 lines to extract the negative sentiment 
    
    score3 = score["neg"]
    score_neg.append(score3)
```

```python
score_com  = []
score_pos = []
score_neg = []

for i in range( 0 , amazon.shape[0]):  ## amazon.shape[0] is for [0]shape means for all rows and to select all rows
    
    score = sent.polarity_scores(amazon.iloc[i][1]) ## i which is already given for rows and 1 is for second column ( review )
    
    ## below two line to extract compound values from the dictionary
    
    score1 = score["compound"]     ## saving compound scores in score1 and append score1 in score_coum
    score_com.append(score1)
    
    ## below 2 lines to extract the positive sentiment 
    
    score2 = score["pos"]
    score_pos.append(score2)
    
    
    ## below two lines are to extract the negative sentiments values 
    
    score3 = score["neg"]
    score_neg.append(score3)
    
```
